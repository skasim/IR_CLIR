```python
import string
def transliterate(hindi):
    s = ""
    for idx, ch in enumerate(hindi):
        if ch in hi2rm:
            s += hi2rm[ch]
        elif ch in string.punctuation:
            s += ch
        elif ch.isdigit():
            s += ch
        elif ch == " ":
            s += ch
        elif ch == "़":
            last_char = s[-1]
            s = s[:-1]
            if last_char == "j":
                s += "z"
            if last_char == "k":
                s += "q"
#             if last_char == "dh":
#                 s += "rh"
#             if last_car == "kh":
#                 s += "kh"
            if last_char == "g":
                s += "gh"
            if last_char == "f":
                s += "f"
    return s

#     "ज़": "z",
#     "क़": "q",
#     "ड़": "r",
#     "ढ़":"rh",
#     "ख़": "kh",
#     "ग़": "gh",
#     "फ़": "f",
```
